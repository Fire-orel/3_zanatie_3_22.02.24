## Task 6

Узнай средний доход среди 'Data Scientist' специалистов и средний доход среди 'Machine Learning/ MLops Engineer' \
Т.к. в колонке Q29 указан промежуток, то возьми среднее значение из промежутка. 

Например: 
- '15,000-19,999' - 17500, 
- '125,000-149,999' - 137500. 
- \> $1,000,000 - оставляем 1000000

Средние доход **округли до целого.**

```python
# Код тут
data_scientist=csv[csv["Q23"]=="Data Scientist"]
machine_learning=csv[csv["Q23"]=="Machine Learning/ MLops Engineer"]
price_data=[]
price_machine=[]
for i in data_scientist["Q29"]:
    try:
        # print(i)
        if i[0]==">":
            i=i.replace(">","")
            i=i.replace("$","")
            price_data.append(float(i.replace(",","")))
        elif i[0]=="$":
            start,end = i.split("-")
            start= start.replace("$","")
            price_data.append((float(start.replace(",",""))+float(end.replace(",","")))/2)
        else:
            start,end = i.split("-")
            price_data.append((float(start.replace(",",""))+float(end.replace(",","")))/2)
    except TypeError:
        pass
for i in machine_learning["Q29"]:
    try:
        # print(i)
        if i[0]==">":
            i=i.replace(">","")
            i=i.replace("$","")
            price_machine.append(float(i.replace(",","")))
        elif i[0]=="$":
            start,end = i.split("-")
            start= start.replace("$","")
            price_machine.append((float(start.replace(",",""))+float(end.replace(",","")))/2)
        else:
            start,end = i.split("-")
            price_machine.append((float(start.replace(",",""))+float(end.replace(",","")))/2)
    except TypeError:
        pass




# print(price_data)
print(len(data_scientist),len(machine_learning))
print(np.round(sum(price_data)/len(price_data),2),np.round(sum(price_machine)/len(machine_learning),2))
```
